### 1. Конвертация из PDF в JPG

Самое рациональное  - с помощью PyMuPDF.

Не требует дополнительных установок на локальный компьютер как напр. poppler в pdf2jpg. Нужно только установить модуль.

Функция перевода всех страниц PDF в JPG

```python
import os
import fitz

def pdf2jpg(filepath: str):
    path = os.path.dirname(filepath)
    basename = os.path.basename(filepath)
    filename = os.path.splitext(basename)[0]
    pdfdoc = fitz.open(filepath)
    for page in pdfdoc: # сохраняет все страницы pdf файла как jpg с номерами страницы в конце имени jpg файла в тот же каталог
        pixmap = page.get_pixmap()
        img = pixmap.tobytes()
        pixmap.save(os.path.join(path, filename)+'_'+str(page.number)+'.jpg') # если нужно сохранить jpg в другой каталог - можно здесь поменять исп. path
```

Пример получения JPG

```python
pdf2jpg('Садовая 29_Часть1.pdf')
```

### 2. Конвертация из JPG в PDF

Самое простое - с помощью fpdf

```python
!pip install --upgrade fpdf
```

Функция сбора JPG файлов в один PDF

```python
from fpdf import FPDF
import glob

def jpg2pdf_collect(filename: str, output: str):
    pdf = FPDF()
    image_files = sorted(glob.glob(filename+'*.jpg')) # сбор всех jpg файлов, начинающихся на filename в текущем каталоге
    for image in image_files:
        pdf.add_page()
        pdf.image(image, 0, 0, 210, 297) # указываем формат выходной страницы A4 210x297мм
    pdf.output(output+'.pdf', 'F') # записываем в переданное имя pdf файла
```

Сбор jpg файлов в 'output.pdf'

```python
jpg2pdf_collect('Садовая 29_Часть1', 'output')
```

### 3. Проверка реального формата файла

Функция проверки, действительно ли файл формата .jpg или .png или .pdf

```python
!pip install python-magic
import imghdr
import magic

# возвращает формат файла
def check_file_format(file_path: str) -> str:
    jpg_format = imghdr.what(file_path)
    if jpg_format == 'jpeg':
        return 'jpeg'
    elif jpg_format == 'png':
        return 'png'

    mime = magic.Magic(mime=True)
    file_mime_type = mime.from_file(file_path)
    if file_mime_type == 'application/pdf':
        return 'pdf'
```

Проверка

```python
print(check_file_format('Садовая 29_Часть1.pdf'))
```
